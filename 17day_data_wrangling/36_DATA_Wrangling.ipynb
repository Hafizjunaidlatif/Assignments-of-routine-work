# Data Normalization
- 1 Uniform the data
- 2 Making user they have same impact
- 3 Eik machli sumandar mn ha or eik jar mn
- also for computational reasons

```python
kashti.head()
```

```python
ks4= kashti[['age_in_days','fare']]
ks4.head()
```

- The above data is really in a wide range and we need to normalize and hard to compare 
- Normaliztion change the values to the range of 0-1 (now both variables has similar influence on our model)

Method of Normalziation

1. Simple feature scaling
    - x(new)=x(old)/x(max)
2. Min-Max method
3. Z-score (standard score) -3 to +3
4. Log transformation

```python
# Simple feature scalling
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4.head()
```

```python
# Simple feature scalling
ks4['age_in_days'] = ks4['age_in_days']/ks4['age_in_days'].max()
ks4.head()
```

```python
# above both function in same code
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4['age_in_days'] = ks4['age_in_days']/ks4['age_in_days'].max()
ks4.head()
```

```python
# MIN-MAX method
ks4['fare']= (ks4['fare']-ks4['fare'].min()) / (ks4['fare'].max()-ks4['fare'].min())
ks4.head()

# its range fall from -1 to 1 (-1,0,1)
```

```python
# Z-score (standard score)
ks4['fare'] = (ks4['fare']-ks4['fare'].mean()) / ks4['fare'].std()
ks4.head()

# its range fall from -3 to 3
```

```python
# log transformation
ks4['fare'] = np.log(ks4['fare'])
ks4.head()
```

```python
# above our data is already very small so ,, but we can use log to normalize real data. now we can try this.
# we will add "ks" as a new fresh dataset
ks = sns.load_dataset('titanic')
ks.head() 
```
