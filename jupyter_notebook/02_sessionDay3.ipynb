##2-Import libararies to be used 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##3-Set theme

```python
sns.set_theme(style="ticks",color_codes=True)
```

##4-Import datasets as you can load your own data

```python
kashti=sns.load_dataset("titanic")
#print(kashti)
```

##5-Plot basic graph with one variable countplot

```python
p=sns.countplot(x="sex",data=kashti)
plt.show()
```

##6-Plot basic graph with two variable countplot

```python
p=sns.countplot(x="sex",hue="class",data=kashti)
plt.show()
```

##7--Plot basic graph with two variable countplot with titles

```python
p=sns.countplot(x="sex",hue="class",data=kashti)
p.set_title("dataviz learning by aammar")
plt.show()
```
